```python
ax = dados.plot(x='Unidade da Federação', y='2008/Ago', kind='bar', figsize=(11,7))
import matplotlib.ticker as ticker
plt.title('Distribuição de verbas entre as federações')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.show()
```
