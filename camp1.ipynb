```python
ax = dados.plot(x='Unidade da Federação', y='2008/Ago', kind='bar', figsize=(11,7))
import matplotlib.ticker as ticker 
plt.title('Distribuição de verbas entre as federações')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.show()
```

```python
ax = dados.plot(x='Unidade da Federação', y='2008/Mar', kind='bar', figsize=(11,7))
import matplotlib.ticker as ticker
plt.title('Distribuição de verbas entre as federações')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.show()
```

```python
ax = dados.plot(x='Unidade da Federação', y='2008/Nov', kind='bar', figsize=(11,7))
import matplotlib.ticker as ticker
plt.title('Distribuição de verbas em Reais entre as federações')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.show()
```

```python
dados_sort = dados[["Unidade da Federação", "2008/Mar"]].sort_values(by = "2008/Mar", ascending = False)
ax = dados_sort.plot(x="Unidade da Federação", y = "2008/Mar", kind ="bar", figsize=(10,6), legend=None, color="purple",)
plt.title('DISTRIBUIÇÃO DE VERBA EM REAIS POR FEDERAÇÃO EM MARÇO/08')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.style.use('dark_background')
plt.show()
####DESAFIO 1 E 2.
```
